```python
import yaml

data_yaml_path = '/content/drive/MyDrive/SA1 DATASCI 3/data.yaml'

data = {
    'train': '/content/drive/MyDrive/SA1 DATASCI 3/train/images',
    'val': '/content/drive/MyDrive/SA1 DATASCI 3/valid/images',
    'test': '/content/drive/MyDrive/SA1 DATASCI 3/test/images',
    'nc': 10,
    'names': ['bike', 'bus', 'car', 'ebike', 'jeepney', 'motorcycle', 'sidecar', 'tricycle', 'truck', 'van']
}

with open(data_yaml_path, 'w') as f:
    yaml.dump(data, f)

print("data.yaml updated!")
print(yaml.dump(data))
```

```python
from ultralytics import YOLO

model1 = YOLO('yolo11n.pt')

results1 = model1.train(
    data=data_yaml_path,
    epochs=25,
    imgsz=640,
    optimizer='AdamW',
    batch=4,
    lr0=0.01,
    name='model1_adamw',
    project='/content/drive/MyDrive/SA1 DATASCI 3/runs'
)
```

```python
model2 = YOLO('yolo11n.pt')

results2 = model2.train(
    data=data_yaml_path,
    epochs=30,
    imgsz=640,
    optimizer='SGD',
    batch=20,
    lr0=0.001,
    name='model2_sgd',
    project='/content/drive/MyDrive/SA1 DATASCI 3/runs'
)
```

```python
model3 = YOLO('yolo11n.pt')

results3 = model3.train(
    data=data_yaml_path,
    epochs=40,
    imgsz=640,
    optimizer='auto',
    batch=-1,
    lr0=0.0001,
    name='model3_auto',
    project='/content/drive/MyDrive/SA1 DATASCI 3/runs'
)
```

```python
from IPython.display import Image, display
import os

for model_name in ['model1_adamw', 'model2_sgd', 'model3_auto']:
    cm_path = f'/content/drive/MyDrive/SA1 DATASCI 3/runs/{model_name}/confusion_matrix.png'
    if os.path.exists(cm_path):
        print(f"\nConfusion Matrix — {model_name}")
        display(Image(cm_path))
    else:
        print(f"Not found: {cm_path}")
```

```python
from ultralytics import YOLO
import pandas as pd

models = {
    'Model 1 (AdamW, batch=4, lr=0.01, epochs=25)': f'/content/drive/MyDrive/SA1 DATASCI 3/runs/model1_adamw/weights/best.pt',
    'Model 2 (SGD, batch=20, lr=0.001, epochs=30)': f'/content/drive/MyDrive/SA1 DATASCI 3/runs/model2_sgd/weights/best.pt',
    'Model 3 (auto, batch=-1, lr=0.0001, epochs=40)': f'/content/drive/MyDrive/SA1 DATASCI 3/runs/model3_auto/weights/best.pt',
}

results_list = []
for model_name, weights_path in models.items():
    model = YOLO(weights_path)
    metrics = model.val(data=data_yaml_path, imgsz=640, verbose=False)

    precision = metrics.box.mp
    recall = metrics.box.mr
    map50 = metrics.box.map50
    f1 = 2 * (precision * recall) / (precision + recall + 1e-9)

    results_list.append({
        'Model': model_name,
        'mAP50': round(map50, 4),
        'Precision': round(precision, 4),
        'Recall': round(recall, 4),
        'F1 Score': round(f1, 4)
    })

metrics_df = pd.DataFrame(results_list)
print("\nModel Performance Metrics")
display(metrics_df)
```
